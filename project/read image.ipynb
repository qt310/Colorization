```python
import glob
import cv2
import matplotlib.pyplot as plt
import random
import numpy as np
from skimage.transform import resize
from scipy.misc import imresize
%matplotlib inline
```

```python
# read images
img_all = []
for img in glob.glob(r"../../images/person/*.jpg"):
    n= cv2.imread(img)#BGR
    if not n is None and np.mean(n)<250 and np.mean(np.std(n,axis=2))>2:
    # drop None pic, almost white pic, almost gray scale pic
        m = imresize(n,(400,400,3))
        img_all.append(m)
```

```python
# show images
import random
import numpy as np
img_list = np.arange(len(img_all))
random.shuffle(img_list)
j = 0
plt.figure(figsize=(15,15))
for i in img_list[:100]:
    j = j+1
    plt.subplot(10,10,j)
#     img = cv2.cvtColor((255*img_all[i]).astype(np.uint8), cv2.COLOR_BGR2RGB)#use resize
    img = cv2.cvtColor(img_all[i], cv2.COLOR_BGR2RGB)#use imresize
    plt.imshow(img)
    plt.axis('off')
    plt.title(i)
#     print(img_all[i].shape)
```
