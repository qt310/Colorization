```python
import glob
import cv2
import matplotlib.pyplot as plt
import random
import numpy as np
from skimage.transform import resize
```

```python
# read images
cv_img = []
for img in glob.glob(r"../images/plant/*.jpg"):
    n= cv2.imread(img)
    if not n is None and np.mean(n)<250:
        m = resize(n,(400,400,3))
        cv_img.append(m)
```

```python
# show images
import random
import numpy as np
img_list = np.arange(len(cv_img))
random.shuffle(img_list)
j = 0
plt.figure(figsize=(15,15))
for i in img_list[:100]:
    j = j+1
    plt.subplot(10,10,j)
    plt.imshow(cv_img[i])
    plt.axis('off')
#     print(cv_img[i].shape)
```
