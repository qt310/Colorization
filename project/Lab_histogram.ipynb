```python
import glob
import cv2
import matplotlib.pyplot as plt
import random
import numpy as np
from skimage.transform import resize
from scipy.misc import imresize
%matplotlib inline
```

```python
img_all = []
for img in glob.glob(r"../../images/person/*.jpg"):
    n= cv2.imread(img)#BGR
    if not n is None and np.mean(n)<250 and np.mean(np.std(n,axis=2))>2:
    # drop None pic, almost white pic, almost gray scale pic
        m = imresize(n,(400,400,3))
        img_all.append(m)
```

```python
edge = np.arange(10,225,10)
j = 1
plt.figure(figsize=(20,15))
for i in range(12):
    img = img_all[i]
    plt.subplot(4,6,j)
    j += 1
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
    img1_lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
    plt.subplot(4,6,j)
    j += 1
    h = plt.hist2d(img1_lab[:,:,1].flatten(),img1_lab[:,:,2].flatten(),bins=edge)
```
